```python
# 数据集下载

import os
import requests
import zipfile

def download_and_extract():
    if os.path.exists('data/tiny-imagenet-200'):
        print("数据集已存在，跳过下载。")
        return

    print("正在下载 Tiny-ImageNet (可能需要几分钟)...")
    url = "http://cs231n.stanford.edu/tiny-imagenet-200.zip"
    r = requests.get(url, stream=True)
    with open("tiny-imagenet-200.zip", "wb") as f:
        for chunk in r.iter_content(chunk_size=1024):
            if chunk: f.write(chunk)

    print("正在解压...")
    with zipfile.ZipFile("tiny-imagenet-200.zip", 'r') as zip_ref:
        zip_ref.extractall("data/")

    print("完成！数据位于 data/tiny-imagenet-200")

download_and_extract()
```
